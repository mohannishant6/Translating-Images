```python
def dodge(front,back):
    result=front*255/(255-back) 
    result[result>255]=255
    result[back==255]=255
    return result.astype('uint8')

def grayscale(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

```

### sketchify and resize and blur 

```python
source_img='backup/real_landscapes'
dest_img='backup/real_sketch/'
blur_sigma=5

for i,image_name in enumerate(os.listdir(source_img)):
    print (image_name)
    s = imageio.imread(os.path.join(source_img,image_name))
    s = cv2.resize(s, (500,500), interpolation = cv2.INTER_AREA)
    g=grayscale(s)
    g = ndimage.gaussian_filter(g, sigma=blur_sigma)
    i = 255-g
    b = scipy.ndimage.filters.gaussian_filter(i,sigma=10)
    r= dodge(b,g)
#     plt.imshow(r, cmap="gray")

    plt.imsave(os.path.join(dest_img,image_name), r, cmap='gray', vmin=0, vmax=255)
```

### concat sketch and real image 

```python
# FROM1=os.path.join('backup','real_sketch')
# FROM2=os.path.join('backup','real_resized')
# TO=os.path.join('backup','joined')

# FROM1=os.path.join('backup','painted_landscapes_train')
# FROM2=os.path.join('backup','real_landscapes_train')
# TO=os.path.join('backup','join_train')

FROM1=os.path.join('backup','subjects_r')
FROM2=os.path.join('backup','subjects_r')
TO=os.path.join('backup','subjects_j')

def join_images(FROM1,FROM2,TO):
    for name in os.listdir(FROM1):
        sketch=os.path.join(FROM1,name)
        real=os.path.join(FROM2,name)
        print(name)

        images = [Image.open(x) for x in [real,sketch]]
#         print (images)
        widths, heights = zip(*(i.size for i in images))

        total_width = sum(widths)
        max_height = max(heights)

        new_im = Image.new('RGB', (total_width, max_height))

        x_offset = 0
        for im in images:
            new_im.paste(im, (x_offset,0))
            x_offset += im.size[0]

        new_im.save(os.path.join(TO,name[:-5]+'.jpg'))
        print(os.path.join(TO,name[:-5]+'.jpg'))
    
join_images(FROM1,FROM2,TO)
```
